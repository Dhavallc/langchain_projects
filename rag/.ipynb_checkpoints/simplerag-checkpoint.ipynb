```python
# web based loader
from langchain_community.document_loaders import WebBaseLoader
from bs4.filter import SoupStrainer

## LOAD CHUNK html

loader = WebBaseLoader(
    web_paths=["https://lilianweng.github.io/posts/2023-06-23-agent/"],
    bs_kwargs=dict(parse_only=SoupStrainer(class_=("post-title", "post-header", "post-content"))),
    requests_kwargs={"headers":{"User-Agent": "Mozilla/5.0 (compatible; MyBot/1.0; +http://example.com/bot)"}}
)
text_documents = loader.load()
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
splitter = RecursiveCharacterTextSplitter(chunk_size = 500,chunk_overlap = 100)
chunks = splitter.split_documents(text_documents)
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 1000,chunk_overlap = 300)
documents = text_splitter.split_documents(text_documents)
documents[:2]
```

```python
from langchain_ollama import OllamaEmbeddings
embeddings = OllamaEmbeddings(
    model = "mxbai-embed-large:latest"
)

#Looping over each doc in text_documents
#Extracting doc.page_content — but only if it exists (i.e., not None or empty)
contents = [doc.page_content for doc in chunks if doc.page_content]
#to do the embeddings
vectors = embeddings.embed_documents(contents)
```

```python
# Assume you already have:
# - vectors: list of embeddings
# - chunks: list of Document objects (same length as vectors)

# Create payloads for Pinecone
pinecone_vectors = [
    {
        "id": f"chunk-{i}",
        "values": vectors[i],
        "metadata": {
            **chunks[i].metadata,  # original metadata from your WebBaseLoader
            "text": chunks[i].page_content,  # snippet preview (optional)
            "chunk_index": i
        }
    }
    for i in range(len(vectors))
]

# Now upsert to Pinecone (assumes `index` is your Pinecone Index object)
index.upsert(vectors=pinecone_vectors)

print(f"✅ Successfully sent {len(pinecone_vectors)} chunks to Pinecone.")
```

```python
import getpass
import os
from pinecone import Pinecone

if not os.getenv("PINECONE_API_KEY"):
    os.environ["PINECONE_API_KEY"]  = getpass.getpass("ENTER YOUR PINECONE KEY:")

pinecone_api_key = os.environ.get("PINECONE_API_KEY")
#pinecone client to converse with pineconce
pc = Pinecone(api_key=pinecone_api_key)
```

```python
from pinecone import ServerlessSpec

index_name = 'langchain-test-db'

if not pc.has_index(index_name):
    print("it is not present need to create index")
    pc.create_index(
        name=index_name,
        dimension=1024,
        metric="cosine",
        spec=ServerlessSpec(cloud='aws',region='us-east-1'),
    )
else:
    print("It is present")
index = pc.Index(index_name)
```

```python
query = "How do autonomous agents use reflection?"
query_vector = embeddings.embed_query(query)

results = index.query(vector=query_vector, top_k=5, include_metadata=True)

for match in results["matches"]:
    print(match["score"], match["metadata"].get("text", ""))
```
